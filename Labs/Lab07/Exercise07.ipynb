### Estimating parameters of a logistic model

Gelman's book has an example where the dose of a drug may be affected to the number of rat deaths in an experiment.

| Dose (log g/ml) | # Rats | # Deaths |
|-----------------|--------|----------|
| -0.896          | 5      | 0        |
| -0.296          | 5      | 1        |
| -0.053          | 5      | 3        |
| 0.727           | 5      | 5        |

We will model the number of deaths as a random sample from a binomial distribution, where $n$ is the number of rats and $p$ the probabbility of a rat dying. We are given $n = 5$, but we believve that $p$ may be related to the drug dose $x$. As $x$ increases the number of rats dying seems to increase, and since $p$ is a probability, we use the following model:

$$
y \sim \text{Bin}(n, p) \\
\text{logit}(p) = \alpha + \beta x \\
\alpha \sim \mathcal{N}(0, 5) \\
\beta \sim \mathcal{N}(0, 10)
$$

where we set vague priors for $\alpha$ and $\beta$, the parameters for the logistic model.

```python
log_reg_code = """
data {
    int<lower=0> n;
    int<lower=0> j;
    vector[j] x; // This is a vector, which contains real by default
    int y[j]; // This is to declare an int array (not vector)
}
transformed data {}
parameters {
    real alpha;
    real beta;
}
transformed parameters {}
model {
    alpha ~ normal(0, 5);
    beta ~ normal(0, 10);
    y ~ binomial_logit(n, alpha + beta * x);
}
generated quantities {}
"""

log_reg_dat = {
             'n': 5,
             'x': np.array([-0.896, -0.296, -0.053, 0.727]),
             'y': np.array([0, 1, 3, 5]),
             'j': 4
            }

fit = pystan.stan(model_code=log_reg_code, data=log_reg_dat, iter=2000, chains=1)
```

### Using a hierarchcical model

This uses the Gelman radon data set and is based off this [IPython notebook](http://twiecki.github.io/blog/2014/03/17/bayesian-glms-3/). Radon levels were measured in houses from all counties in several states. Here we want to know if the preence of a basement affects the level of radon, and if this is affected by which county the house is located in. 

The data set provided is just for the state of Minnesota, which has 85 counties with 2 to 116 measurements per county. We only need 3 columns for this example `county`, `log_radon`, `floor`, where `floor=0` indicates that there is a basement.

We will perfrom simple linear regression on log_radon as a function of county and floor.


#### Hiearchical model

With a hierarchical model, there is an $a_c$ and a $b_c$ for each county $c$ just as in the individual couty model, but they are no longer indepnedent but assumed to come from a common group distribution

$$
a_c \sim \mathcal{N}(\mu_a, \sigma_a^2) \\
b_c \sim \mathcal{N}(\mu_b, \sigma_b^2)
$$

we furhter assume that the hyperparameters come from the following distributions

$$
\mu_a \sim \mathcal{N}(0, 100^2) \\
\sigma_a \sim \mathcal{U}(0, 100) \\ 
\mu_b \sim \mathcal{N}(0, 100^2) \\
\sigma_b \sim \mathcal{U}(0, 100)
$$

```python
radon = pd.read_csv('radon.csv')[['county', 'floor', 'log_radon']]
radon.head()
```

```python
county = pd.Categorical(radon['county'])
# Will use country.labels to get the numerical code for county
```

```python
hier_code = """
data {
    int<lower=0> N;
    int<lower=1,upper=85> county[N];
    vector[N] floornum;
    vector[N] lradon;
}
transformed data {}
parameters {
    real mu_a;
    real sig_a;
    real mu_b;
    real sig_b;
    vector[85] a;
    vector[85] b;
    real sig_y;
}
transformed parameters {
    vector[N] y_hat;
    for (i in 1:N) {
        y_hat[i] <- a[county[i]] + b[county[i]] * floornum[i];
    }
}
model {
    mu_a ~ normal(0, 100^2); // mean of a
    mu_b ~ normal(0, 100^2); // mean of b
    sig_a ~ uniform(0, 100); // Variance of a
    sig_b ~ uniform(0, 100); // Variance of b
    
    sig_y ~ gamma(10, 1); // Household error
    
    lradon ~ normal(y_hat, sig_y);
}
generated quantities {}
"""

hier_dat = {
             'N': len(radon),
             'floornum': radon['floor'],
             'lradon': radon['log_radon'],
             'county': county.labels + 1
            }

fit = pystan.stan(model_code=hier_code, data=hier_dat, iter=2000, chains=1)
```

**Exercise 3 (10 points)**: Implement the leap-frog method and simulate the spring dynamics following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the spring trajectory plots.

```python
from JSAnimation import IPython_display
from matplotlib import animation
```

```python
def U(x):
    return x ** 2 / 2.0
def K(p):
    return p ** 2 / 2.0
def dU(x):
    return x
def dK(p):
    return p
```

```python
delta = 0.1
L = 70

# Initial position
x0 = -4
p0 = 1

pSteps = [p0]
xSteps = [x0]

# First half step
pStep = p0 - delta / 2.0 * dU(x0)
# First full step
xStep = x0 + delta * dK(pStep)
for i in range(L - 1):
    pStep = pStep - delta * dU(xStep)
    xStep = xStep + delta * dK(pStep)
    pSteps.append(pStep)
    xSteps.append(xStep)
# Last half step
pStep = pStep - delta / 2.0 * dU(xStep)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211, xlim=(-6, 6), ylim=(-1, 1), 
                      title='Harmonic Oscillator')
line1, = ax1.plot([], [], 'k-')
ball1, = ax1.plot([], [], 'bo', linewidth=40)
ax2 = fig.add_subplot(223, xlim=(-0.5, 1.5), ylim=(0, 20), 
                      title='Energy')
line2, = ax2.plot([], [], color='b')
line3, = ax2.plot([], [], color='y')
line4, = ax2.plot([], [], color='r')

ax3 = fig.add_subplot(224, xlim=(-5, 5), ylim=(-5, 5), aspect='equal',
                      xlabel='x', ylabel='p', title="Phase Space")
ball2, = ax3.plot([], [], 'o', linewidth=40)

fig.subplots_adjust(hspace=0.5)
plot_objs = [line1, ball1, line2, line3, line4, ball2]

def init():
    for obj in plot_objs:
        obj.set_data([], [])
    return plot_objs

def animate(i):
    xx = np.linspace(-6, xSteps[i], 1000)
    y = np.sin(6*np.linspace(0, 2*np.pi, 1000))
    line1.set_data(xx, y)
    ball1.set_data(xSteps[i], 0)
    
    u = U(xSteps[i])
    p = K(pSteps[i])
    line2.set_data([0, 0], [0, u])
    line3.set_data([0, 0], [u, u + p])
    line4.set_data([1, 1], [0, u + p])
    
    ball2.set_data(xSteps[i], pSteps[i])
    return plot_objs

animation.FuncAnimation(fig, animate, init_func=init, frames=L, interval=70, blit=True)
```

**Exercise 4 (20 points)**: Now implement the HMC following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the bivariate normal MCMC plots. (For another simple explanation of HMC, see Chapter 14 of "Doing Basyesian Data Analysis" by Jouh Kruschke.) 

```python
# Bivariate normal parameters
Sigma = np.array([[1, .8], [.8, 1]])

# Potential energy function
def U(x):
    return x.T.dot(np.linalg.inv(Sigma)).dot(x)
def dU(x): # Gradient of potential energy
    return x.T.dot(np.linalg.inv(Sigma))

# Kinetic energy function
def K(p):
    return p.dot(p) / 2.0
```

```python
np.random.seed(12345)

# Step size
delta = 0.3
nSamples = 1000
L = 20

# Initial state
x = np.zeros((2, nSamples))
x0 = np.array([0, 6])
x[:,0] = x0

t = 0
while t < (nSamples-1):
    t = t + 1
    
    # Sample random momentum
    p0 = np.random.standard_normal(2)

    ## Simulate Hamiltonian momentum
    # First 1/2 step
    
    pStar = p0 - delta / 2.0 * dU(x[:, t-1])
    
    # First full step for position / sample
    xStar = x[:, t-1] + delta * pStar
    
    # Full steps
    for f in range(L-1):
        # Momentum
        pStar = pStar - delta * dU(xStar)
        # Position / Sample
        xStar = xStar + delta * pStar
    
    # Last half step
    pStar = pStar - delta / 2.0 * dU(xStar)
    
    # Evaluate energies at start and end of trajectory
    U0 = U(x[:, t-1])
    UStar = U(xStar)
    
    K0 = K(p0)
    KStar = K(pStar)
    
    # Acceptance / rejection criterion
    alpha = np.min([1, np.exp((U0 + K0) - (UStar + KStar))])
    
    u = np.random.uniform()
    if u < alpha:
        x[:,t] = xStar
    else:
        x[:,t] = x[:,t-1]
```

```python
plt.plot(x[0, :], x[1, :], 'k.')
plt.plot(x[0, 0:50], x[1, 0:50], 'ro-', linewidth=2)
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.legend(['Samples', '1st 50 states'])
plt.title('Hamilton Monte Carlo')
```
