```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
%matplotlib inline
```

```python
from JSAnimation import IPython_display
from matplotlib import animation
```

```python
def U(x):
    return x ** 2 / 2.0
def K(p):
    return p ** 2 / 2.0
def dU(x):
    return x
def dK(p):
    return p

delta = 0.1
L = 70

# Initial position
x0 = -4
p0 = 1

pSteps = [p0]
xSteps = [x0]

# First half step
pStep = p0 - delta / 2.0 * dU(x0)
# First full step
xStep = x0 + delta * dK(pStep)
for i in range(L - 1):
    pStep = pStep - delta * dU(xStep)
    xStep = xStep + delta * dK(pStep)
    pSteps.append(pStep)
    xSteps.append(xStep)
# Last half step
pStep = pStep - delta / 2.0 * dU(xStep)
```

```python
# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(-6, 6), ylim=(-1, 1))
line, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
def animate(i):
#     x = np.linspace(0, 2, 1000)
#     y = np.sin(2 * np.pi * (x - 0.01 * i))
    x = np.linspace(-6, xSteps[i], 1000)
    y = np.sin(6*np.linspace(0, 2*np.pi, 1000))
    line.set_data(x, y)
    return line,

# call the animator.  blit=True means only re-draw the parts that have changed.
animation.FuncAnimation(fig, animate, init_func=init,
                        frames=L, interval=50, blit=True)
```
